## 1. Environment Setup

```python
import requests
import json
from typing import Optional, Dict
from pathlib import Path
from IPython.display import Audio, display

# AOG service configuration
AOG_BASE_URL = "http://localhost:16688"
AOG_API_SPEC_VERSION = "v0.2"
DEFAULT_MODEL = "NamoLi/speecht5-tts"

print("âœ… Configuration complete")
print(f"ğŸ“ AOG service address: {AOG_BASE_URL}")
print(f"ğŸ¤– Default model: {DEFAULT_MODEL}")
print("âš ï¸  Note: Currently only supports English text")
```

## 2. Helper Functions

```python
def call_aog_tts(
    text: str,
    model: str = DEFAULT_MODEL,
    voice: str = "male"
) -> Optional[Dict]:
    """
    Call AOG Text-to-Speech API
    
    Args:
        text: éœ€è¦è½¬æ¢æ–‡æœ¬(ä»…æ”¯æŒè‹±æ–‡)
        model: Model name
        voice: éŸ³è‰² (male/female/girl/baby)
    
    Returns:
        API response data
    """
    try:
        url = f"{AOG_BASE_URL}/aog/{AOG_API_SPEC_VERSION}/services/text-to-speech"
        
        payload = {
            "model": model,
            "text": text,
            "voice": voice
        }
        
        print(f"ğŸ“¤ Sending request...")
        print(f"ğŸ“ æ–‡æœ¬: {text[:50]}..." if len(text) > 50 else f"ğŸ“ æ–‡æœ¬: {text}")
        print(f"ğŸ¤ éŸ³è‰²: {voice}")
        
        response = requests.post(url, json=payload, timeout=60)
        response.raise_for_status()
        
        result = response.json()
        
        if result.get("business_code") == 200:
            print("âœ… Speech synthesis successful")
            audio_path = result.get("data", {}).get("url")
            if audio_path:
                print(f"ğŸ”Š éŸ³é¢‘æ–‡ä»¶: {audio_path}")
        else:
            print(f"âš ï¸  ä¸šåŠ¡é”™è¯¯: {result.get('message')}")
        
        return result
        
    except requests.exceptions.ConnectionError:
        print("âŒ æ— æ³•è¿æ¥åˆ° AOG æœåŠ¡")
        return None
    except requests.exceptions.Timeout:
        print("âŒ Request timeout")
        return None
    except Exception as e:
        print(f"âŒ é”™è¯¯: {e}")
        return None

def play_audio_file(audio_path: str):
    """
    åœ¨ notebook ä¸­Play Audioæ–‡ä»¶
    
    Args:
        audio_path: éŸ³é¢‘æ–‡ä»¶è·¯å¾„
    """
    if not audio_path:
        print("âš ï¸  æ²¡æœ‰éŸ³é¢‘æ–‡ä»¶å¯æ’­æ”¾")
        return
    
    if not Path(audio_path).exists():
        print(f"âš ï¸  éŸ³é¢‘æ–‡ä»¶ä¸å­˜åœ¨: {audio_path}")
        return
    
    try:
        print(f"\nğŸ”Š Play Audio: {audio_path}")
        display(Audio(audio_path, autoplay=False))
    except Exception as e:
        print(f"âŒ æ’­æ”¾å¤±è´¥: {e}")

print("âœ… Helper Functionså®šä¹‰å®Œæˆ")
```

## 3. Basic Speech Synthesis

```python
# åŸºæœ¬ç¤ºä¾‹
text = "Hello, this is a text to speech example. Welcome to AOG!"

response = call_aog_tts(text)

if response and response.get("business_code") == 200:
    audio_path = response.get("data", {}).get("url")
    if audio_path:
        play_audio_file(audio_path)
```

## 5. Batch Generation

ç”Ÿæˆå¤šæ®µè¯­éŸ³

```python
# Batch Generationç¤ºä¾‹
texts = [
    "Good morning!",
    "How are you today?",
    "Have a nice day!"
]

audio_files = []

for i, text in enumerate(texts, 1):
    print(f"\nğŸ¤ ç”Ÿæˆç¬¬ {i}/{len(texts)} æ®µè¯­éŸ³...")
    response = call_aog_tts(text)
    
    if response and response.get("business_code") == 200:
        audio_path = response.get("data", {}).get("url")
        if audio_path:
            audio_files.append(audio_path)

print(f"\nâœ… æˆåŠŸç”Ÿæˆ {len(audio_files)}/{len(texts)} æ®µè¯­éŸ³")

# æ’­æ”¾æ‰€æœ‰ç”ŸæˆéŸ³é¢‘
for audio_path in audio_files:
    play_audio_file(audio_path)
```
