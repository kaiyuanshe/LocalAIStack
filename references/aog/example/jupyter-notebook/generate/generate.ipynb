## 1. Environment Setup

```python
import requests
import base64
import json
from typing import List, Optional, Dict
from pathlib import Path

# AOG service configuration
AOG_BASE_URL = "http://localhost:16688"
AOG_API_SPEC_VERSION = "v0.2"
DEFAULT_MODEL = "gemma3:4b"

print("âœ… Configuration complete")
print(f"ğŸ“ AOG service address: {AOG_BASE_URL}")
print(f"ğŸ¤– Default model: {DEFAULT_MODEL}")
```

## 2. Helper Functions

```python
def call_aog_generate(
    prompt: str,
    model: str = DEFAULT_MODEL,
    stream: bool = False,
    think: bool = False,
    images: Optional[List[str]] = None
) -> Optional[Dict]:
    """
    Call AOG Generate API
    
    Args:
        prompt: Text prompt
        model: Model name
        stream: æ˜¯å¦æµå¼Returns
        think: æ˜¯å¦Returnsæ€è€ƒè¿‡ç¨‹
        images: List of base64 encoded images
    """
    try:
        url = f"{AOG_BASE_URL}/aog/{AOG_API_SPEC_VERSION}/services/generate"
        
        payload = {
            "model": model,
            "prompt": prompt,
            "stream": stream,
            "think": think
        }
        
        if images:
            payload["images"] = images
        
        print(f"ğŸ“¤ Sending request...")
        response = requests.post(url, json=payload, timeout=60)
        response.raise_for_status()
        
        print("âœ… Generation successful")
        return response.json()
        
    except requests.exceptions.ConnectionError:
        print("âŒ æ— æ³•è¿æ¥åˆ° AOG æœåŠ¡")
        return None
    except Exception as e:
        print(f"âŒ é”™è¯¯: {e}")
        return None

def image_to_base64(image_path: str) -> Optional[str]:
    """å°†å›¾åƒæ–‡ä»¶è½¬æ¢ä¸º base64 ç¼–ç """
    try:
        with open(image_path, "rb") as f:
            return base64.b64encode(f.read()).decode()
    except Exception as e:
        print(f"âŒ å›¾åƒç¼–ç å¤±è´¥: {e}")
        return None

def print_response(response: Optional[Dict]):
    """Format and print response"""
    if not response:
        return
    
    if response.get("success") and "message" in response:
        content = response["message"].get("response", "")
        print("\n" + "="*60)
        print("ğŸ¤– Generation result:")
        print("="*60)
        print(content)
        print("="*60 + "\n")
    else:
        print("âš ï¸  Abnormal response format")
        print(json.dumps(response, indent=2, ensure_ascii=False))

print("âœ… Helper Functionså®šä¹‰å®Œæˆ")
```

## 3. Pure Text Generation

```python
# Pure Text Generationç¤ºä¾‹
prompt = "Write a short poem about spring"

response = call_aog_generate(prompt)
print_response(response)
```
